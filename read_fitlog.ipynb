```python
import pandas as pd
import numpy as np
```

```python
fittingdata = pd.read_csv('/Users/rosaliaobrien/research/GALFIT_CLEAR/running_GALFIT/fitting_data.csv')
outputdata = pd.read_csv('/Users/rosaliaobrien/research/GALFIT_CLEAR/running_GALFIT/output_data.csv')
outputdata
```

```python
errordf = pd.DataFrame([])

for flter in ['160', '125', '850']:

    for gid in outputdata['gid']:
        
        field = fittingdata.loc[fittingdata['gid'] == gid, 'field'].item()

        if field == 'GND':
            fullfield = 'North'
        if field == 'GSD':
            fullfield = 'South'

        print('\n')
        print('\n')
        print(gid, field, flter)

        fitlogpath = '/Users/rosaliaobrien/research/GALFIT_CLEAR/running_GALFIT/'+fullfield+'/'+flter+'/fit.log'
        with open(fitlogpath) as fitlog:
            fitlog = fitlog.readlines()

        #read feedme file for specific gid
        with open('/Users/rosaliaobrien/research/GALFIT_CLEAR/running_GALFIT/'+fullfield+'/'+flter+'/outputs/'+str(gid)+'/'+str(gid)+'.feedme') as feedmefile:
            feedme = feedmefile.readlines()

        #make list of gids based on readme file
        gidlist_from_files = []
        for index, line in enumerate(feedme):
            try:
                if line.split()[0] == '#ID:':
                    gidlist_from_files.append(int(line.split()[1]))

            except:
                pass

        #set fitting_object_index as the index from gidlist_from_files
        for k, j in enumerate(gidlist_from_files):
            if j == gid:
                print(j,k)
                fitting_object_index = k

        print('fitting_object_index = {}'.format(fitting_object_index))
        print('feedme gidlist: {}'.format(gidlist_from_files))

        #run through every line of fitlog looking for the LATEST instance of that gid
        for i in range(0,len(fitlog)):

            if str(gid) in fitlog[-i]:
                
                k = fitting_object_index

                print(fitlog[-i+2*k+3])
                print(fitlog[-i+2*k+3].split())

                #latest instance will be in line like:
                #Output image    : 27006.fits
                #errors will be 3 lines ABOVE this line
                #if there are multiple objects in a single fit, add 2*k to the number of lines above the output image line to get the correct errors
                
                if fitlog[-i+2*k+3].split()[0] == '(':
                    xpos = fitlog[-i+2*k+3].split()[1].split(',')[0]
                    ypos = fitlog[-i+2*k+3].split()[2].split(')')[0]
                    mag = fitlog[-i+2*k+3].split()[3]
                    rad = fitlog[-i+2*k+3].split()[4]
                    sersic = fitlog[-i+2*k+3].split()[5]
                    ar = fitlog[-i+2*k+3].split()[6]
                    PA = fitlog[-i+2*k+3].split()[7]
                    
                if not fitlog[-i+2*k+3].split()[0] == '(':
                    xpos = fitlog[-i+2*k+3].split()[0].split(',')[0].split('(')[0]
                    ypos = fitlog[-i+2*k+3].split()[1].split(')')[0]
                    mag = fitlog[-i+2*k+3].split()[2]
                    rad = fitlog[-i+2*k+3].split()[3]
                    sersic = fitlog[-i+2*k+3].split()[4]
                    ar = fitlog[-i+2*k+3].split()[5]
                    PA = fitlog[-i+2*k+3].split()[6]
                    
                #return a string of the actual calculated data so that you can confirm that you are getting the correct errors
                calcdata = fitlog[-i+2*k+2]

                refit = 'no'
                
                if '*' in fitlog[-i+2*k+3]:
                    refit = 'yes'
                    print(refit)
                    
                if 'nan' in fitlog[-i+2*k+3]:
                    refit = 'yes'

                errordf = errordf.append(pd.DataFrame([[gid,field,flter,xpos,ypos,mag,rad,sersic,ar,PA,refit,calcdata]], columns = ['gid', 'field', 'filter', 'x_err', 'y_err', 'mag_err','rad_err','sersic_err','ar_err','PA_err','refit','calcdata']), ignore_index = True)

                break
                
errordf.to_csv('/Users/rosaliaobrien/research/GALFIT_CLEAR/running_GALFIT/GALFIT_errors.csv')
```
