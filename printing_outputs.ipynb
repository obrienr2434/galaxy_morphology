```python
#Objects from Vince's paper

gids = [23758, 37955, 16758, 43615, 42221, 39241, 45972, 44620, 
        39631, 39170, 34694, 23435, 47677, 39805, 38785, 32566, 
        40476, 21156, 17070, 35774, 40597, 37686, 46066, 40862, 
        39804, 21427, 40623, 41520, 40223, 39012, 44042]

fields = ['N', 'N', 'N', 'S', 'S', 'S', 'S', 'S',
          'S', 'S', 'N', 'N', 'S', 'S', 'S', 'N', 
          'S', 'N', 'N', 'S', 'S', 'N', 'S', 'S', 
          'S', 'N', 'S', 'S', 'S', 'S', 'S']
```

```python
#Separate GOODS N and GOODS S

gidsn =[]
gidss =[]

for field, gid in zip(fields, gids):
    if field == 'N':
        gidsn.append(gid)
        
    else:
        gidss.append(gid)
```

```python
def plot(field, gids, flter):
    
    if not (field == 'S' or field == 'N'):
        raise Exception('Field must be S or N.')
    
    if field == 'N':
        fullfield = 'North'
        
    if field =='S':
        fullfield = 'South'
        
    if not type(flter) == str:
        raise Exception('Filter must be string.')
        
    os.chdir('/Users/rosaliaobrien/research/science_project/gal_morphology/'+fullfield+'/'+flter+'/')

    for i, gid in enumerate(gids):

        file = 'N/fits/'+str(gid)+'.fits'
        savefig = 'N/pngs/output_gid'+str(gid)+'.png'

        if not os.path.exists('N/pngs'):
                os.mkdir('N/pngs')

        image = fits.getdata(file, ext=1)
        model = fits.getdata(file, ext=2)
        residual = fits.getdata(file, ext=3)

        fig = plt.figure(figsize=(20, 7))

        gs = gridspec.GridSpec(1, 3, wspace=0.01, hspace=0)

        ax1 = plt.subplot(gs[0,0])
        ax2 = plt.subplot(gs[0,1], sharey = ax1)
        ax3 = plt.subplot(gs[0,2], sharey = ax1)

        ax1.set_title('HST Image Object '+str(gid))
        ax2.set_title('GALFIT Model')
        ax3.set_title('Residual')

    #     ax1.set_xlim(left=170, right=250)
    #     ax2.set_xlim(left=170, right=250)
    #     ax3.set_xlim(left=170, right=250)

    #     ax1.set_ylim(bottom=170, top=300)

        ax1.set_xticklabels([])
        ax1.set_yticklabels([])
        ax2.set_xticklabels([])
        ax2.set_yticklabels([])
        ax3.set_xticklabels([])
        ax3.set_yticklabels([])

        ax1.imshow(image, vmin=0, vmax=0.02, cmap='Greys_r', origin='lower', aspect="auto")
        ax2.imshow(model, vmin=0, vmax=0.02, cmap='Greys_r', origin='lower', aspect="auto")
        ax3.imshow(residual, vmin=0, vmax=0.02, cmap='Greys_r', origin='lower', aspect="auto")

        plt.savefig(savefig)

        plt.show()
```

```python
plot('N', gidsn, '160')
```
